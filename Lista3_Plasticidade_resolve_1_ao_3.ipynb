```python
import math
import numpy as np
# Dados de Experimento 1
h_1= float(input("Digite a umidade do corpo de prova do ensaio Nº1: \n\n"))
N_1 = int(input("Digite de golpes do ensaio Nº1: \n\n"))
# Calculando LL usando a fórmula
LL_1 = (h_1/100)/ (1.419 - (0.3 * math.log10(N_1)))

# Imprimindo o valor de LL
print(f"Limite de Liquidez (LL)do ensaio Nº1 é de: {LL_1:.3f} \n\n")
```
